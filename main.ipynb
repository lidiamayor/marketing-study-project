# Data-Analysis marketing strategy for a wine company

We are a marketing firm specialized in data analysis and digital strategies, recently hired by a renowned wine company to optimize their sales through data-driven insights. Our goal is to thoroughly analyze the available data to craft the most effective marketing campaign.

We will focus on three key areas:

**Identifying the largest consumer group**: By examining demographic factors such as age, gender, and income, we will pinpoint who the top wine consumers are. This will enable us to create highly targeted and personalized campaigns for the right audience.

**Understanding the most popular purchasing channels**: We will assess whether customers prefer to buy in physical stores, through catalogs, or online. This information will help direct marketing efforts toward the most effective sales channels.

**Optimizing ad click-through rates**: By analyzing consumer interests (fashion, technology, travel, sports, etc.), we will identify which types of ads generate the highest engagement and clicks, especially across different income groups. This will allow us to create compelling ads that resonate with the target audience.

## Marketing Study

```python
from functions_marketing import clean_df_marketing, site_purchases_by_age, site_purchases_by_income, web_visits_by_age, income_by_ages, purchases_by_income, purchases_by_income_line, purchases_by_education, son_at_home, purchases_by_living_status, purchases_by_month

url = 'datasets/marketing_campaign.xlsx'
df = pd.read_excel(url)
```

```python
df = clean_df_marketing(df)
```

```python
df_wine = df[df['MntWines']>200]
```

```python
site_purchases_by_age(df_wine)
```

The table shows that **store purchases** are consistently the most popular across all age groups. Younger consumers (16-24) prefer **catalog and web purchases**, with relatively low **deals purchases**. The **35-44** age group stands out with the highest **deals and web purchases**, indicating a stronger interest in promotions and online shopping. As age increases, **catalog purchases** become more common, especially among those **65-74**, while **deals and web purchases** slightly decrease. Overall, store shopping remains the dominant choice across all ages.

```python
site_purchases_by_income(df_wine)
```

Lower income groups (20k-40k) make the most deals and have high web purchases, but fewer catalog purchases.
Middle income groups (40k-60k) have a balance of web and store purchases, with moderate deals and catalog purchases.
Higher income groups (60k-80k and 80k-100k) see a decline in deals and a rise in catalog purchases, with high store purchases.
Highest income group (100k-120k) makes no deals but has the highest catalog and store purchases.

```python
web_visits_by_age(df_wine)
```

This table shows the average monthly web visits by age range. The most active group online is those aged 35-44, with nearly 5 visits per month. Web visits gradually decline in older age groups, particularly for those over 65, who average 3.49 visits. Younger groups (16-24 and 25-34) have the lowest averages, with just over 3 visits per month, indicating less frequent web activity compared to middle-aged customers.

```python
income_by_ages(df)
```

This table shows that average income increases with age. The youngest group (16-24) has the lowest average income at 51,813, while the income peaks at 56,814 for those aged 55-64. Incomes are higher in older age ranges, with a notable rise from the 25-34 group to the 45-54 group.
